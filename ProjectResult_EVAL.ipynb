```python
import torch
from scipy.spatial.distance import cosine
from transformers import AutoModel, AutoTokenizer



def cos_sim(embeddings=None):
  for i in range(len(embeddings)):
    for j in range(len(embeddings)):
        if i != j:
          print('cosine difference between',text[i],'and',text[j],'is',1 - cosine(embeddings[i],embeddings[j])) 


def run_sentence_similarity_scores(model_path=None, sentence_list = None):
  tokenizer = AutoTokenizer.from_pretrained(model_path)
  model = AutoModel.from_pretrained(model_path)

  if sentence_list :
    texts = sentence_list
  else:
    ''
  inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

  with torch.no_grad():
    embeddings = model(**inputs, output_hidden_states=True, return_dict=True).pooler_output

  return embeddings;

#text to use for comparison
text = [
    "I love machine learning",
    "language models are great",
    "the cat in the hat"
]
```

### **Test 1. Searching similar sentence using BERT embeddings**

---

```python
model_path = "bert-base-uncased"
embeds = run_sentence_similarity_scores(model_path = model_path, sentence_list = text);
cos_sim(embeds)
```

### **Test 2. Searching similar sentences using base simCSE**

---

```python
model_path = "princeton-nlp/sup-simcse-bert-base-uncased"
embeds = run_sentence_similarity_scores(model_path = model_path, sentence_list = text);
cos_sim(embeds)
```
